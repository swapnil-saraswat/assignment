```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sp.csv")
df
```

```python
sns.relplot(x='race/ethnicity',y='parental level of education',data=df,hue='gender',style='lunch',col='test preparation course')
#this shows two graphs of test preparation(1:none,2:completed).
# on x-axis groups are distributed
# on y-axis parental level of education is given
#this graph shows how many male female done which degree with which lunch
```

```python
sns.relplot(x='writing score',y='reading score',data=df,hue='gender',col='test preparation course',style='lunch')
#this represent positive relation between writing and reading score
```

```python
sns.relplot(x='race/ethnicity',y='parental level of education',data=df,hue='gender',style='lunch',col='test preparation course',kind= "line")
```

```python
sns.catplot(x='gender',y='math score',data=df,kind='bar')
#this bar graph shows that relation between gender and their math score
```

```python
sns.barplot(x='parental level of education',y='math score',data=df,hue='gender',)
plt.xticks(rotation=90)
#this group shows in which dgree scoring in math of male and female
```

```python
df['parental level of education'].value_counts()
```

```python
label=list(df['parental level of education'].value_counts().keys())
value =list(df['parental level of education'].value_counts().values)
```

```python
ls=[0,0.2,0,0,0,0.2]
sns.set(rc={'figure.figsize':(6,6)})
plt.pie(value,labels=label,wedgeprops={'ec':'k'},explode=ls,autopct='%1.2f%%',)#ec=edge ,k=color
plt.title('parental level of education')
plt.show()
#this shows pie chart of the level of education which conclude only 5.90% engaged in master's degree
```
